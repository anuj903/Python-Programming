# matplotlib Tutorial

```python
from matplotlib import pylab
print (pylab.__version__)
```

```python
# use Numpy to generate random Data

import numpy as np
x = np.linspace(0,10,25)
y = x * x+2

print(x)
print(y)
print(np.array([x,y]).reshape(25,2))
```

```python
# it only takes one command to draw

pylab.plot(x,y,'r')
# r stands for red
```

```python
# Drawing a subgraph

pylab.subplot(1,2,1) # the contents of the brackets represent (rows, columns, indexes)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,2) #(1,2,1)
pylab.plot(y,x,'g*--')
```

```python
# (1,2,1) and (1,2,2) are significant if we want titled subplots, without which both the graphs will 
# be in the same plot
# example

pylab.subplot(1,2,1) # the contents of the brackets represent (rows, columns, indexes)
pylab.plot(x,y,'r--')
pylab.subplot(1,2,1) #(1,2,1)
pylab.plot(y,x,'g--')
```

```python
from matplotlib import pyplot as plt
```

```python
fig = plt.figure()
axes = fig.add_axes([0.5,0.1,0.8,0.8]) # control the left, right, wdth, height of the frames of the canvas (from 0 to 1)

axes.plot(x,y,'r')
```

```python
# Subplots

fig, axes = plt.subplots(nrows = 1, ncols = 2) # submap is of 1 row, 2 columns
for ax in axes :
    ax.plot(x,y,'r')
```

```python
# Draw a picture , or graph inside another graph
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) # big axes

axes2 = fig.add_axes([0.2,0.6,0.2,0.2]) # small canvas

axes1.plot(x,y,'r')
axes2.plot(y,x,'g')
```

# Other 2D Graphs

```python
n = np.array([0,1,2,3,4,5])

fig, axes = plt.subplots(1,4, figsize =(16,5))
axes[0].set_title("scatter plot")
axes[0].scatter(x,x+0.25*np.random.randn(len(x)))

axes[1].set_title("step")
axes[1].step(n,n**2,lw=2)

axes[2].set_title("bar")
axes[2].bar(n,n**2, align ="center",alpha=0.5)

axes[3].set_title("Fill Between")
axes[3].fill_between(x,x**2,x**3, color ="green",alpha=0.5)
```

```python
# Radar Chart

fig = plt.figure(figsize=(6,6))
ax = fig.add_axes([0.0,0.0,1,1],polar = True)
t = np.linspace(0,2 * np.pi,100)
ax.plot(t,t*2,color = 'blue',lw =3)
```

```python
# histogram

n = np.random.randn(100000)
fig,axes = plt.subplots(1,2, figsize=(12,4))
axes[0].set_title("histogram")
axes[0].hist(n)
```

```python

n = np.random.randn(100000)
fig,axes = plt.subplots(1,2, figsize=(12,4))
axes[0].set_title("histogram")
axes[0].hist(n)

axes[1].set_title("Cumulative histogram")
axes[1].hist(n, cumulative = True, bins= 10)
```

```python
# Contour Map
```

```python
import matplotlib
import numpy as np
import matplotlib.cm as cm
import matplotlib.pyplot as plt

delta = 0.025
x = np.arange(-3.0,3.0,delta)
y = np.arange(-2.0,2.0,delta)
X,Y = np.meshgrid(x,y)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X - 1)**2 - (Y - 1)**2)
Z = (Z1 -Z2) * 2

print(X)
print(Y)
```

```python
fig, ax = plt.subplots()
CS = ax.contour(X,Y,Z)
ax.clabel(CS, inline=1, fontsize=10)

ax.set_title('contour map')
```
